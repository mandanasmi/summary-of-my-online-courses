```python
import pandas as pd
```

```python
diabetes = pd.read_csv('pima-indians-diabetes.csv')
```

```python
diabetes.head()
```

```python
diabetes.columns
```

```python
colls_to_norm = ['Number_pregnant', 'Glucose_concentration', 'Blood_pressure', 'Triceps',
       'Insulin', 'BMI', 'Pedigree']
```

```python
# feature normalization

diabetes[colls_to_norm] = diabetes[colls_to_norm].apply(lambda x: (x-x.min())/ (x.max()-x.min() ))
```

```python
diabetes.head()
```

```python
import tensorflow as tf
```

```python
diabetes.columns
```

```python
num_preg = tf.feature_column.numeric_column('Number_pregnant')
plasma_gluc = tf.feature_column.numeric_column('Glucose_concentration')
dias_press = tf.feature_column.numeric_column('Blood_pressure')
tricep = tf.feature_column.numeric_column('Triceps')
insulin = tf.feature_column.numeric_column('Insulin')
bmi = tf.feature_column.numeric_column('BMI')
pedigree = tf.feature_column.numeric_column('Pedigree')
age = tf.feature_column.numeric_column('Age')
```

```python
assigned_gp = tf.feature_column.categorical_column_with_vocabulary_list('Group',['A','B','C','D'])
```

```python
#other alternative: assigned_gp = tf.feature_column.categorical_column_with_hash_bucket('Group',hash_bucket_size=10)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diabetes['Age'].hist()
```

```python
age_bucket = tf.feature_column.bucketized_column(age, boundaries=[20,30,40,50,60,70,80]) 
# continuos numeric feature to categorical column
```

```python
feat_cols = [num_preg, plasma_gluc,dias_press, tricep, insulin, bmi,pedigree, assigned_gp, age_bucket]
```

```python
#Train_Test_split
```

```python
x_data = diabetes.drop('Class', axis=1) #only the features
```

```python
x_data.head()
```

```python
labels = diabetes['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, labels, test_size = 0.3, random_state=101)
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train, y=y_train, batch_size=10, num_epochs=1000,shuffle=True)
```

```python
model = tf.estimator.LinearClassifier(feature_columns=feat_cols,n_classes=2)
```

```python
model.train(input_fn=input_func,steps=1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test, y=y_test,batch_size=10,
                                                     num_epochs=1,
                                                     shuffle = False)
```

```python
results = model.evaluate(eval_input_func)
```

```python
results
```

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test,batch_size=10,num_epochs=1, shuffle=False)
```

```python
predictions = model.predict(pred_input_func) # prediction 
```

```python
list(predictions)
```

# Dense Neural Network Classifier 

```python
dnn_model = tf.estimator.DNNClassifier(hidden_units=[10,10,10], feature_columns=feat_cols,n_classes=2) 
# 3 layer, each layer 10 nodes
```

```python
dnn_model.train(input_fn=input_func,steps=1000) # this is giving us an error because we need to passing 

#categorical columns to an embedding column 
```

```python
embedded_group_col = tf.feature_column.embedding_column(assigned_gp, dimension=4) 
#dimension of group which is A,B,C,D here
```

```python
feat_cols = [num_preg, plasma_gluc,dias_press, tricep, insulin, bmi, 
             pedigree, embedded_group_col, age_bucket]
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(X_train, y_train,batch_size=10, num_epochs=1000, shuffle=True )
```

```python
dnn_model = tf.estimator.DNNClassifier(hidden_units=[10,20,20,20,10], feature_columns=feat_cols, n_classes=2)
```

```python
dnn_model.train(input_fn=input_func, steps=1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test, y=y_test, batch_size=10, num_epochs=1, shuffle=False)
```

```python
dnn_model.evaluate(eval_input_func)
```
