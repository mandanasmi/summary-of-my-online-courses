# TF Estimator API

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_data = np.linspace(0.0,10.0,1000000)
```

```python
noise = np.random.randn(len(x_data))
```

```python
y_true = (0.5 * x_data) + 5 + noise
```

```python
feat_cols = [ tf.feature_column.numeric_column('x',shape=[1]) ]
```

```python
estimator = tf.estimator.LinearRegressor(feature_columns=feat_cols)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_eval, y_train, y_eval = train_test_split(x_data, y_true, test_size = 0.3, random_state=101)
```

```python
print(x_train.shape)
```

```python
x_eval.shape
```

```python
#define input function
input_func = tf.estimator.inputs.numpy_input_fn({'x':x_train},y_train,batch_size=8,num_epochs=None,shuffle=True)
```

```python
#for evaluation 
train_input_func = tf.estimator.inputs.numpy_input_fn({'x':x_train},y_train,batch_size=8,num_epochs=1000,shuffle=False)
```

```python
eval_input_func = tf.estimator.inputs.numpy_input_fn({'x':x_eval},y_eval,batch_size=8,num_epochs=1000,shuffle=False)
```

```python
estimator.train(input_fn=input_func, steps=1000)
```

```python
train_metrics = estimator.evaluate(input_fn=train_input_func,steps=1000)
```

```python
eval_metrics = estimator.evaluate(input_fn=eval_input_func,steps=1000)
```

```python
print('Training data metrics')
print(train_metrics)
```

```python
print('Evaluation  metrics')
print(eval_metrics)
```

```python
# When the loss of train and test data metric are almost the same (we want them to be as close as possible) 
# we can figure out our model is not overfitting to our training data
```

```python
brand_new_data = np.linspace(0,10,10)
```

```python
input_fn_predict = tf.estimator.inputs.numpy_input_fn({'x':brand_new_data},shuffle=False)
```

```python
list(estimator.predict(input_fn=input_fn_predict)) # generator object: need to iterate through this or cast it to a list
```

```python
preditions = []

for pred in estimator.predict(input_fn=input_fn_predict):
    preditions.append(pred['predictions'])
```

```python
preditions
```
