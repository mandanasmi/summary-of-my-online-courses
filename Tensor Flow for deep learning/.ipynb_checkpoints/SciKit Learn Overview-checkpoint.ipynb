```python
import numpy as np
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data = np.random.randint(0,100,(10,2))
```

```python
data
```

```python
scaler_model = MinMaxScaler() 
```

```python
type(scaler_model)
```

```python
scaler_model.fit(data)
```

```python
scaler_model.transform(data)
```

```python
scaler_model.fit_transform(data) # perform both 2 steps in one step
# usually we fit only the train data and then transform train and test data both 
```
